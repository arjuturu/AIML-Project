```python
from collections import defaultdict
from collections import Counter

#customer names
customerNamesList = ["Amar","Kalpana","Anwika","Ramu","Raju","Jeevan"];
#print(customerNamesList)
allCustomerOrders = [
    ("Amar", "Fridge", 112, "Home Essentials"),
    ("Kalpana", "Phone", 563, "Electronics"),
    ("Anwika", "Frock", 5, "Clothing"),
    ("Ramu", "Laptop", 732, "Electronics"),
    ("Jeevan", "Phone", 665, "Electronics"),
    ("Raju", "Pant", 18, "Clothing"),
    ("Amar", "Fridge", 135, "Home Essentials"),
    ("Kalpana", "Smartphone", 619, "Electronics"),
    ("Anwika", "Dress", 16, "Clothing"),
    ("Ramu", "Laptop", 698, "Electronics"),
    ("Raju", "Jeans", 10, "Clothing"),
    ("Jeevan", "Washing Machine", 247, "Home Essentials"),
    ("Amar", "Microwave", 95, "Home Essentials"),
    ("Kalpana", "Headphones", 33, "Electronics"),
    ("Anwika", "Skirt", 10, "Clothing"),
    ("Ramu", "Tablet", 281, "Electronics"),
    ("Raju", "T-Shirt", 7, "Clothing"),
    ("Jeevan", "Air Conditioner", 394, "Home Essentials"),
    ("Amar", "Smartwatch", 135, "Electronics"),
    ("Kalpana", "Saree", 28, "Clothing"),
    ("Anwika", "Mixer Grinder", 45, "Home Essentials"),
    ("Ramu", "Bluetooth Speaker", 50, "Electronics"),
    ("Raju", "Kurta", 8, "Clothing"),
    ("Jeevan", "Vacuum Cleaner", 87, "Home Essentials"),
    ("Amar", "LED TV", 338, "Electronics"),
    ("Kalpana", "Blouse", 13, "Clothing"),
    ("Rana", "Blouse", 13, "Clothing")
]


# Use a dictionary where keys are customer names and values are lists of ordered products
customersOrderedProducts = {}

for name, product, price, category in allCustomerOrders:
    if name not in customersOrderedProducts:
        customersOrderedProducts[name] = []
    customersOrderedProducts[name].append((product, price, category))
print("\nProducts by Customer:")
print(customersOrderedProducts)

#2. Classify products by category
productsByCategroy = {}
for name, product, price, category in allCustomerOrders:
    if category not in productsByCategroy:
        productsByCategroy[category] = []
    if product not in productsByCategroy[category]:
        productsByCategroy[category].append(product)
print("\nproductsByCategroy : ")
print(productsByCategroy)

# 3. Analyze customer orders
customerByValue = {}
for key,value in customersOrderedProducts.items():
    totalAmount = 0
    for order in value:
        totalAmount = totalAmount + order[1]
        if totalAmount > 100:
            customerByValue[key]=(totalAmount,"High-Value Buyer")
        elif totalAmount > 50 and totalAmount < 100:
            customerByValue[key]=(totalAmount,"Moderate-Value Buyer")
        else:
            customerByValue[key]=(totalAmount,"Low-Value Buyer")
print("\nCustomers By Value(High,Moderate and Low")
print(customerByValue)

print("\nGenerate business insights")

#4.1 Calculate the total revenue per product category and store it in a dictionary
totalRevenueByCategory = {}
for orders in customersOrderedProducts.values():
    for order in orders:
          category = order[2]  # Extract category from the order
          price = int(order[1])  # Convert price to integer
          # Accumulate total revenue per category
          totalRevenueByCategory[category] = totalRevenueByCategory.get(category, 0) + price
print("\n4.1 totalRevenueByCategory")
print(totalRevenueByCategory)

# • Extract unique products from all orders using a set
uniqueProducts = {}
for orders in customersOrderedProducts.values():
    for order in orders:
        uniqueProducts[order[0]] = uniqueProducts.get(order[0], 0) + 1
print("\n4.2 Unique Prodcuts ordered by customers")
# for key,value in uniqueProducts.items():
#     if value == 1:
#         print(f"{key} : {value}")
print(sorted(uniqueProducts))
# • Use a list comprehension to find all customers who purchased electronics
electronicsPurchasedCustomers = [item for item in allCustomerOrders if item[3] == "Electronics"]
print("\n4.2 customers who purchased electronics")
print(electronicsPurchasedCustomers)

# • Identify the top three highest-spending customers using sorting
for key,value in customerByValue.items():
    topThreeCustomers = sorted(customerByValue.items(), key=lambda x: x[1][0], reverse=True)[:3]
print("\ntop three highest-spending customers: ")
print(topThreeCustomers)

# • Print a summary of each customer’s total spending and their classification
print("\ncustomer’s total spending and their classification:")
print(customerByValue)

# • Use set operations to find customers who purchased from multiple categories
customersPurchasedFromMultipleCategories = {}
categorySet = set()
for order in allCustomerOrders:
    if order[0] not in customersPurchasedFromMultipleCategories:
        customersPurchasedFromMultipleCategories[order[0]] = set()
    customersPurchasedFromMultipleCategories.get(order[0]).add(order[3])
print('\ncustomers who purchased from multiple categories')
for key,value in customersPurchasedFromMultipleCategories.items():
    if(len(value) > 1):
         print(f"{key}: {value}")
# • Identify common customers who bought both electronics and clothing
required = {"Clothing", "Electronics"}
print('\nCustomer who bought Electronics and Clothing')
for key,value in customersPurchasedFromMultipleCategories.items():
    if required.issubset(value):
         print(f"{key}: {value}")
        
#Identify trends based on category-wise sales      
category_product_count = defaultdict(lambda: defaultdict(int))
for customer, product, price, category in allCustomerOrders:
    category_product_count[category][product] += 1
# Display results
for category, products in category_product_count.items():
    print(f"\nCategory: {category}")
    for product, count in sorted(products.items()):
        print(f"  - {product}: {count} time(s)")

# Most frequently purchased products
product_list = [order[1] for order in allCustomerOrders]

product_frequency = Counter(product_list)
most_frequent_products = product_frequency.most_common(5)
print("Most Frequently Purchased Products:")
for product, count in most_frequent_products:
    print(f"- {product}: {count} time(s)")
```
